$s$: CARを計算できたPRの文書集合、$s_1 (\subset s)$: CARが高かったPRの文書集合、$s_2 (:= s - s_1)$: CARが高くなかったPRの文書集合。単語$t$に対して、それぞれ以下のように定義。

$$
tf(t, s_1) = \frac{df(t, s_1)}{|s_1|}
$$

$$
idf(t) = \log\frac{|s|}{df(t,s)}
$$

$$
df(t, x) = \left|\{d\mid d\ni t, d\in x\}\right|
$$

$$
tf{\rm -}idf(t, s_1) = tf(t, s_1)\cdot idf(t)
$$

tf-idfが高い単語から並べる。

```python
pr_mor_fname = "pr_mor"
car_fname = "car.new"
```

```python
from collections import defaultdict

car_all = defaultdict(list)
with open(car_fname, 'r') as f:
    for line in f:
        articleid, prtype, code, val = line.strip().split('\t')
        car_all[articleid].append(float(val) * 100.)
        
car = {}
for articleid, val_list in car_all.items():
    car[articleid] = sum(val_list) / len(val_list)
    
#car
```

```python
# わかち書きデータの読み込み

pr_mor = {}
with open(pr_mor_fname, 'r') as f:
    for line in f:
        articleid, mor = line.strip().split('\t')
        if articleid in car:
            pr_mor[articleid] = mor.split(' ')

print(pr_mor['NIKPRLRSP038060_07012003'])
```

```python
s = [set(mor) for articleid, mor in pr_mor.items()]
n_s = len(s)
```

```python
car_th = 10.
s1 = [set(mor) for articleid, mor in pr_mor.items() if car[articleid] >= car_th]
n_s1 = len(s)
```
