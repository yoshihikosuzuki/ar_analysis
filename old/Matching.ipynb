```python
def run_jumanpp(s):
    import subprocess
    with open("tmp", 'w') as f:
        f.write(s)
    ret = subprocess.check_output("cat tmp | jumanpp", shell=True).decode('utf-8').strip().split('\n')
    subprocess.call("rm tmp", shell = True)
    return ret
```

```python
# 形態素に分解して、企業名の形態素についてそれぞれ完全マッチを行う
def search_comp_name(pr_entry, lc):
    pr_id, pr_date, pr_body = pr_entry
    pr_split = run_jumanpp(pr_body)
    
    # 形態素だけのリストを生成
    pr_mor = [x.split(' ')[0] for x in pr_split]
    #print(pr_mor)
    
    # pr_splitから住所に関わる部分を抽出？
    
    matched_comp = set()
    for data in lc:
        lc_name, lc_address, lc_code = data
        lc_name_split = run_jumanpp(lc_name)
        lc_address_split = run_jumanpp(lc_address)
        
        lc_name_mor = [x.split(' ')[0] for x in lc_name_split]
        
        while "@" in lc_name_mor:
            lc_name_mor.remove("@")
        while "EOS" in lc_name_mor:
            lc_name_mor.remove("EOS")
            
        count_name = 0
        for n in lc_name_mor:
            #print(n)
            if n in pr_mor:
                #print("hit")
                count_name += 1
        if count_name == len(lc_name_mor):
            matched_comp.add(lc_name)
        print(lc_name, "end")
    print(matched_comp)
```

```python
# 形態素は無視して、企業名で本文から完全マッチ検索
def search_comp_name_exact(pr_entry, lc):
    pr_id, pr_date, pr_body = pr_entry
    
    matched_comp = set()
    for data in lc:
        lc_name, lc_address, lc_code = data
        if lc_name in pr_body:
            matched_comp.add(lc_name)
    print(matched_comp)
```

```python
# 市区町村名だけを抽出
def split_address(adr):
    import subprocess, re
    result = subprocess.check_output("curl http://asp.ncm-git.co.jp/eCapGCWebApi/keyword/geocode/?keyword=\"" + adr.replace(" ", "").replace("　", "") + "\"", shell = True).decode('utf-8')
    ret = set()
    
    def cut_at(num, r):
        import re
        pat = re.compile(r'<authorizedName index="%d">(.*?)<' % num)
        if pat.search(r):
            return pat.search(r).groups()[0]
            
    for i in [2, 3]:
        r = cut_at(i, result)
        if r != None:
            ret.add(r)
            
    return ret
```

```python
lc_city_fname = "listed_company.city"
lc_city = []
with open(data_dir + lc_city_fname, 'r') as f:
    for line in f:
        lc_city.append(set(line.strip().split(' ')))
        
print(lc_city[:5])
```

```python
def do_exact_matching(pr_body, lc, lc_city):
    #print(pr_body)
    
    matched_comp = set()
    for i in range(len(lc)):
        lc_name, lc_address, lc_code = lc[i]
        lc_c = lc_city[i]
        #lc_address_split = split_address(lc_address)
        if lc_name not in pr_body:
            continue
        #else:
        #    print(lc_name, lc_c)
        match_count = 0
        for adr in lc_c:
            if adr in pr_body:
                match_count = 1
        if match_count == 1:
            matched_comp.add(lc_name)
    return matched_comp
```

```python
do_exact_matching(pr[1][2], lc, lc_city)
```

```python
results = []
for i in range(len(pr)):
    results.append((i, do_exact_matching(pr[i][2], lc, lc_city)))
#print(results)
```
